### 합성곱

* 밀집층은 x1*w1+x2*w2... +b(절편)
* 합성공은 뉴런이 가지는 n개의 가중치에 대해 n개씩 절편을 더한다.
* 합성공 신경망에서는 완전 연결 신경망과 달리 뉴런을 **필터 **또는 **커널** 라고 부른다.
* 2차원 배열에서 사용시 필터도 2차원이여야 한다.
  - 왼쪽 위 -> 오른쪽 위 -> 왼쪽 아래 -> 오른쪽 아래

* 합성곱 계산을 통해 얻은 출력을 특별히 **특성 맵** 이라고 부른다.


```python
from tensorflow import keras
keras.layers.Conv2D(10, kernel_size=(3, 3), activation = 'relu') # 순서대로 필터, 필터에 사용할 커널의 크기, 활성화 함수(렐루)
```

* **패딩** : 입력 배열의 주위를 가상의 원소로 채우는 것
* **세임 패딩** : 입력 주위에 0으로 패딩하는 것
* **밸리드 패딩** : 패딩 없이 순수한 입력 배열에서만 합성곱을 하여 특성 맵을 만드는 경우

```python
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding="same") # padding의 기본값은 'valid', 세임 패딩을 사용하려면 'same'
```

* **스트라이드** : n칸씩 이동하면서 만드는 것을
* 기본값 1

```python
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding='same', strides=1)
```

* **폴링** : 합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄이는 역할
* 특성 맵의 크기는 줄이지 않는다.
* 가중치가 없다.
* 도장을 찍은 영역에서 가장 큰 값을 고르거나 평균을 계산 -> **평균 폴링, 최대 폴링**
* MaxPooling2D 클래스로 케라으세어 폴링을 수행할 수 있다.
    * keras.layers.MaxPooling2D(파라미터 값)
    * Ex) keras.layers.MaxPooling2D(2, stides=2, padding='valid')
* AveragePooling2D는 평균 풀링을 제공한다.

* 꼭 기억할 점 : 폴링은 가로세로 방향으로만 진행
